```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pickle
```

```python
dataset=pd.read_csv("PT_INR_Prothrombin _Introduction _Test _Result_Dataset .csv")
```

```python
dataset
```

```python
print(dataset.columns)
```

```python
dataset = dataset.drop(columns=['Patient ID','Health Condition','Treatment Type','Age','Gender','INR Value',])
```

```python
dataset 
```

```python
# Perform one-hot encoding with drop_first=True
dataset= pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
# Convert specific columns from True/False to 0/1
columns = ['Risk Level_Low','Risk Level_Moderate']
dataset[columns] = dataset[columns].astype(int)
```

```python
dataset
```

```python
independent=dataset[["PT Value"]]
independent
```

```python
dependent=dataset[["Risk Level_Low"]]
dependent
```

```python
#SKlearn is library supporting to meachinelearing,sklearn is contain  class of multiplefunctions[model_selection]as class of file[train_test_split]asfunction file
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#sklearn is library function,linear_model is folder to find linear simple regersion(weight,slop(coef)),(bias as a intercept).,regerssor as model variables,linear regerctionis class of all function non parameterized assing
from sklearn.linear_model import LinearRegression
regerssor=LinearRegression()
regerssor.fit(x_train,y_train)
```

```python
#coef_ is weight calculate,regerssor as overall model.
weight=regerssor.coef_
```

```python
weight
```

```python
bias=regerssor.intercept_
```

```python
bias
```

```python
#(cross check),as now creat y_pred variable,overall model in regerssor,AI based on predict so predict as function(x_test),evolution matrix actual value
y_pred=regerssor.predict(x_test)
```

```python
#library function.metrics as a folder import r2_score as file name,r_score is a variable(r2_score)as a function,(y_test,y_pred)as a actual dataset and predict dataset
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#pickel is use to file saved (function),create file dummy file name "finalised_model_linear.sav",(sav),algorithm linear,pickel extension save
import pickle
filename="finalized_model_PT_INR_Prothrombin.sav"
```

```python
pickle.dump(regerssor,open(filename,"wb"))
```

```python
loaded_model=pickle.load(open("finalized_model_PT_INR_Prothrombin.sav","rb"))
result=loaded_model.predict([[18.355144	]])
```
